```python
from PIL import Image, ImageDraw
from math import sqrt, pi, cos, sin
from collections import defaultdict

image = io.imread("4b.tiff",plugin = "tifffile")
for i in range(72):
    print(i)
    img = image[i,:,:,2]
    img = img/np.max(img)
    plt.imshow(img,cmap=plt.cm.gray)
    plt.show()
    med = filters.median(img, disk(4))
    med = med/np.max(med)
    edge = feature.canny(med,sigma=0.1)
    plt.imshow(edge,cmap=plt.cm.gray)
    plt.show()

    canny_result = []

    for r in range(edge.shape[0]):
        for c in range(edge.shape[1]):
            if edge[r,c]:
                canny_result.append((r,c))
    # Find circles
    rmin = 8
    rmax = 15
    steps = 100
    threshold = 0.4

    points = []
    for r in range(rmin, rmax + 1):
        for t in range(steps):
            points.append((r, int(r * cos(2 * pi * t / steps)), int(r * sin(2 * pi * t / steps))))

    acc = defaultdict(int)
    for x, y in canny_result:
        for r, dx, dy in points:
            a = x - dx
            b = y - dy
            acc[(a, b, r)] += 1

    circles = []
    for k, v in sorted(acc.items(), key=lambda i: -i[1]):
        x, y, r = k
        if v / steps >= threshold and all((x - xc) ** 2 + (y - yc) ** 2 > rc ** 2 for xc, yc, rc in circles):
            print(v / steps, x, y, r)
            circles.append((x, y, r))
    
    print("-------")
```
